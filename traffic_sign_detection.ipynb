```python
#filename = os.path.basename(os.path.normpath(path))
for filename in os.listdir("/content/drive/MyDrive/traffic/Test"):
    if filename != ".ipynb_checkpoints":
        f = filename[:-4] + ".xml"
        tree = ET.parse("/content/drive/MyDrive/traffic/Sample_Dataset/annotations/" + f)
        root = tree.getroot()
        #Trafic Light;Stop;Speedlimit;Crosswalk.
        label = root[4][0].text
        nlabel = ""
        if label == "crosswalk":
            nlabel = "Pedestriancrossing"
        else:
            if label == "stop":
                nlabel = "Stop"
            else:
                if label == "speedlimit":
                    nlabel = "Speedlimit"
        for x in root[4][5][:]:
            nlabel += ' %s' % (x.text)
        txtfn = "/content/drive/MyDrive/traffic/Sample_Dataset/groundtruths/" + filename[:-4] + ".txt"
        with open(txtfn, 'w') as f:
            f.write(nlabel)
            f.write('\n')
```
